```python
import os
result = [] 
with open("../apsi_p2/100.MSPEC") as my_file:
    
    dict_info_struct = {"iupac": "", "ion_mode": "", "instrument": "", "instrument_type": "", "ionization": "", "collision_energy": ""
                    , "collision_gas": "", "spectrum_type": "", "precursor_type": "", "precursor_mz": "", "formula": "",
                    "molecular_weight": "", "cas": "", "spectra" : None
                 } 

    current_lines = [] 
    for line in my_file:
        if len(line) == 1:
            dict_info = dict_info_struct.copy() 
            for e in current_lines: 
                
                if "Name" in e: 
                    dict_info['iupac'] = e[6:].strip('\n')

                if "Ion_mode" in e: 
                    dict_info['ion_mode'] = e[10:].strip('\n')

                if "Instrument:" in e:
                    dict_info['instrument'] = e[12:].strip('\n')

                if "Instrument_type" in e: 
                    dict_info['instrument_type'] = e[17:].strip('\n')

                if "Ionization" in e: 
                    dict_info['ionization'] = e[12:].strip('\n')

                if "Collision_energy" in e: 
                    dict_info['collision_energy'] = e[18:].strip('\n')

                if "Collision_gas" in e: 
                    dict_info['collision_gas'] = e[15:].strip('\n')

                if "Spectrum_type" in e: 
                    dict_info['spectrum_type'] = e[15:].strip('\n')

                if "Precursor_type" in e: 
                    dict_info['precursor_type'] = e[16:].strip('\n')

                if "PrecursorMZ" in e: 
                    dict_info['precursor_mz'] = e[13:].strip('\n') 

                if "Formula" in e: 
                    dict_info['formula'] = e[9:].strip('\n')

                if "MW" in e: 
                    dict_info['molecular_weight'] = e[4:].strip('\n') 
                    

                if "ExactMass" in e: 
                    dict_info['exact_mass'] = e[11:].strip('\n') 

                if "CAS" in e: 
                    position = e.find(';')
                    dict_info['cas'] = e[6: position].strip('\n')

                if "Num Peaks" in e: 
                    peak_tuple_result = [] 
                    num_of_peaks = int(e[11:]) 
                    peaks_lines = current_lines[(-1 * num_of_peaks):]
                    for p in peaks_lines: 
                        split_peak_line = p.split() 
                        peak_tuple_result.append((split_peak_line[0], split_peak_line[1]))

                    dict_info['spectra'] = peak_tuple_result
                
            result.append(dict_info) 
            # finished 
            current_lines = [] 
            
                
        current_lines.append(line) 
        
        
            
```
